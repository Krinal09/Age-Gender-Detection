```python
from keras.models import load_model
import os
from os import listdir
from numpy import asarray,save
from keras.utils import load_img, img_to_array
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load gender model save
gender_model= load_model('/content/drive/MyDrive/Model Save/Gender_Detection.h5')
age_model = load_model('/content/drive/MyDrive/Model Save/Age_Detection.h5')
```

```python
# Đường dẫn thư mục test
test="/content/drive/MyDrive/Gender and Age Detection Final Project Data/Age Dataset Main Final Project/Identify_test_age"

# Tạo tên hiện của giới tính và độ tuổi
classes = ['man','woman']
age_classes = ['0-10','11-20','21-30','31-40','41-50','51-60','61-70','71-80','81-90','91-120']
age_classes_reversed = list(reversed(age_classes))
# Quét hết các hình trong đường dẫn và tiến hành dự đoán các loại tiền
for i in os.listdir(test):
  img=load_img(test+'/'+i,target_size=(96,96), color_mode = 'rgb')
  plt.imshow(img)

  img=img_to_array(img)
  img=img.astype('float32')
  img=img/255
  img=np.expand_dims(img,axis=0)
  result=(gender_model.predict(img).argmax())
  result1=(age_model.predict(img).argmax())

  print("Giới tính là {} và độ tuổi là {}".format(classes[result],age_classes_reversed[result1]))
  plt.show()
```
